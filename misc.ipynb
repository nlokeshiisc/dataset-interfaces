# Avoiding empty classes
ImageFolder dataset is causing error when some of the classes in some of the shifts are missing. Now, in such cases, let us assume that the prompt msimply copied over the base prompt as is and did not apply beta to it.

```
Problem cases are snow, forest and rain

Overall we copied 62 classes and the details are available
```
[here](/raid/infolab/nlokesh/dataset-interfaces/attic/imstar_data/copied_classes.txt)

````python
base_dir = Path("/raid/infolab/nlokesh/dataset-interfaces/data/imagenet_star/base")
base_classes = [entry.name for entry in list(base_dir.glob("*"))]
print(len(base_classes))

shifts = constants.ALLSHIFTS

for shift in shifts:
    print(f"Shift: {shift}")

    shift_dir = constants.imagenet_star_dir / shift
    n = 0
    
    for cls in base_classes:
        cls_dir = shift_dir / cls
        num_imgs = len(list(cls_dir.glob("*"))) 

        if num_imgs < 50:
            print(f"{shift} {cls} {num_imgs}")
        else:
            n += 1

    print(f"Num classes: {n}")
````

````python
problem_shifts = [constants.RAIN, constants.SUNLIGHT, constants.FOREST]
base_dir = constants.imagenet_star_dir / constants.BASE
base_classes = [x.name for x in base_dir.iterdir() if x.is_dir()]

num_cpy = 0
for shift in problem_shifts:
    for cls in base_classes:
        shift_dir = constants.imagenet_star_dir / shift / cls   
        if not shift_dir.exists():
            print(f"Creating {shift_dir}")
            # shift_dir.mkdir(parents=True)
        if len(list(shift_dir.glob("*"))) == 0:
            print(f"Copying {base_dir / cls} from base to {shift_dir / cls}")
            shutil.copytree(base_dir / cls, shift_dir / cls)
            num_cpy += 1

    for cls in base_classes:
        shift_dir = constants.imagenet_star_dir / shift / cls   
        assert len(list(shift_dir.glob("*"))) > 0, f"{shift_dir} is empty!"

print(num_cpy)
````
